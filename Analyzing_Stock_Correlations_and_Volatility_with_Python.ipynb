```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the stock symbols and date range
symbols = ['AMD', 'NVDA', 'INTC', 'TSM']
start_date = '2020-01-01'
end_date = '2024-08-17'  # Assuming current date
```

```python
# Fetch the stock data
data = yf.download(symbols, start=start_date, end=end_date)['Adj Close']
```

```python
# Summary statistics for prices
print("Summary Statistics for Prices:\n")
print(data.describe())
```

```python
# Calculate the correlation matrix
correlation_matrix = data.pct_change().corr()
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap of Semiconductor Stocks')
plt.show()
```

```python
# Calculate daily returns
returns = data.pct_change()

# Calculate annualized volatility
volatility = returns.std() * (252 ** 0.5)  # 252 trading days in a year

print("Annualized Volatility:")
print(volatility)
```

```python
# Plot volatility
plt.figure(figsize=(10, 6))
volatility.plot(kind='bar')
plt.title('Annualized Volatility of Semiconductor Stocks')
plt.ylabel('Volatility')
plt.show()
```
